```python
df_zero_mask = df == 0 #fizemos uma variavel para identificar todos os IDs com resultado 0
```

```python
feature_zero_mask = df_zero_mask.iloc[:,1:].all(axis=1) #aqui estamos identificando as linhas com informações do ID que contém 0
#pegamos o zero_mask e colocamos o .iloc[] para pegar todas as linhas. [:(pegar todas as linhas), 1:(examinar as colunas começando pela segunda coluna)]
#.all() ao longo do eixo da coluna, ele retorna TRUE se todas as colunas nessa linha forem true
```

```python
sum(feature_zero_mask) #soma dos booleanos true
```

```python
df_clean_1 = df.loc[~feature_zero_mask,:].copy() #'~'significa "NOT"
```
